```python
#!pip install matplotlib
from deepface import DeepFace  #deepface lib-->DeepFace Class
import cv2 as cv
import matplotlib.pyplot as plt

```

```python
img1=cv.imread("EmotionDetection@MaitriAI/images/Rajpal Yadav.jpeg")
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
result=DeepFace.analyze(img1,actions=['emotion'])
```

```python
print(result)
```

```python
dominant_emotion = result[0]['dominant_emotion']
print(dominant_emotion) 
```
